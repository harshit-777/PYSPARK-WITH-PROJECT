```python
from pyspark.sql import SparkSession 
```

```python
spark = SparkSession.builder.appName('log').getOrCreate()
```

```python
from pyspark.ml.classification import LogisticRegression
```

```python
data_df = spark.read.format('libsvm').option("numFeatures","692").load('data/sample_libsvm_data.txt')
```

```python
data_df.show()
```

```python
log_model = LogisticRegression()
```

```python
log_fit = log_model.fit(data_df)
```

```python
log_summ = log_fit.summary
```

```python
log_summ.predictions.show()
```

```python
lr_train , lr_test = data_df.randomSplit([0.7,0.3])
```

```python
final_mo = LogisticRegression()
```

```python
fit_final = final_mo.fit(lr_train)
```

```python
pred = fit_final.evaluate(lr_test)
```

```python
pred.predictions.show()
```

```python
from pyspark.ml.evaluation import BinaryClassificationEvaluator,MulticlassClassificationEvaluator
```

```python
evaluate_bin = BinaryClassificationEvaluator()
```

```python
my_final = evaluate_bin.evaluate(pred.predictions)
```
